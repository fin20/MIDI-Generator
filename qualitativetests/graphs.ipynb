```python
# Load the CSV file
df = pd.read_csv("/content/drive/MyDrive/finalyearproject/qualitativetests/responses.csv")
# print(df.head())

# List of song columns
song_cols = [f'Song {i}:' for i in range(1, 11)]

# Group by Expertise and calculate the average rating for each song
avg_ratings = df.groupby('Expertise')[song_cols].mean().loc[['Novice','Intermediate','Advanced']]


songs = np.arange(1, 11)  # Song numbers 1 to 10
width = 0.25  # Width of each bar

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot bars for each expertise group
positions_novice = songs - width
positions_intermediate = songs
positions_advanced = songs + width

ax.bar(positions_novice, avg_ratings.loc['Novice'], width, label='Novice')
ax.bar(positions_intermediate, avg_ratings.loc['Intermediate'], width, label='Intermediate')
ax.bar(positions_advanced, avg_ratings.loc['Advanced'], width, label='Advanced')

# Set the labels and title
ax.set_xlabel('Song Number')
ax.set_ylabel('Average Rating (1 = 100% AI, 10 = 100% Human)')
ax.set_title('Responses visualised')
ax.set_xticks(songs)
ax.set_ylim(1, 10)
ax.legend()

plt.show()
```

```python

df = pd.read_csv("/content/drive/MyDrive/finalyearproject/qualitativetests/responses.csv")

# List of song columns
song_cols = [f'Song {i}:' for i in range(1, 11)]


avg_ratings = df.groupby('Expertise')[song_cols].mean().loc[['Novice', 'Intermediate', 'Advanced']]

# Define expertise level
expertise_levels = ['Novice', 'Intermediate', 'Advanced']
colors = ['blue', 'darkorange', 'green']

# Create separate graphs for each expertise level
for i, expertise in enumerate(expertise_levels):
    fig, ax = plt.subplots(figsize=(10, 5))

    # Bar positions (1 to 10)
    songs = np.arange(1, 11)

    # Plot the bars for the specific expertise level
    ax.bar(songs, avg_ratings.loc[expertise], color=colors[i], width=0.5, label=expertise)

    # Set labels and title
    ax.set_xlabel('Song Number')
    ax.set_ylabel('Average Rating (1 = 100% AI, 10 = 100% Human)')
    ax.set_title(f'Responses Visualised - {expertise}')
    ax.set_xticks(songs)
    ax.set_ylim(1, 10)
    ax.legend()

    # Show the plot
    plt.show()
```
